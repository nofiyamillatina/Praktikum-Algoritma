```python
!pip install colorama

from colorama import Style, Fore

# Kelas Barang untuk menyimpan informasi tentang barang
class Barang:
    def __init__(self, kode_barang, nama_barang, harga, jumlah_stok):
        # Inisialisasi atribut barang
        self.kode_barang = kode_barang
        self.nama_barang = nama_barang
        self.harga = harga
        self.jumlah_stok = jumlah_stok

    def info_barang(self):
        # Mengembalikan informasi barang dalam format string
        return f"Kode: {self.kode_barang}, Nama: {self.nama_barang}, Harga: {self.harga}, Stok: {self.jumlah_stok}"

    def update_stok(self, jumlah):
        # Memperbarui jumlah stok dan mencetak pesan sukses
        self.jumlah_stok += jumlah
        print(Style.BRIGHT +
              Fore.GREEN +
              f"Stok barang {self.nama_barang} berhasil diperbarui. Jumlah barang sekarang: {self.jumlah_stok}" +
              Style.RESET_ALL)

# Fungsi untuk menghasilkan kode barang unik berdasarkan jumlah barang yang ada
def generate_kode_barang():
    return str(len(barang_dict) + 1)

# Dictionary untuk menyimpan semua barang dengan kode sebagai kunci
barang_dict = {}

# Fungsi untuk memvalidasi input dari pengguna
def validasi_input(prompt, tipe_data):
    while True:
        input_user = input(prompt).strip()
        if not input_user:
            print(Style.BRIGHT +
                  Fore.RED +
                  "Input tidak boleh kosong." +
                  Style.RESET_ALL)
            continue
        try:
            return tipe_data(input_user)  # Mengubah input sesuai tipe data yang diinginkan
        except ValueError:
            print(Style.BRIGHT +
                  Fore.RED +
                  f"Input harus berupa {tipe_data.__name__}." +
                  Style.RESET_ALL)

# Fungsi untuk menambah barang baru ke dalam inventaris
def tambah_barang():
    while True:
        nama_barang = input("Masukkan nama barang: ").strip()
        if not nama_barang:
            print(Style.BRIGHT +
                  Fore.RED +
                  "Nama barang tidak boleh kosong." +
                  Style.RESET_ALL)
            continue

        harga = validasi_input("Masukkan harga barang: ", float)
        jumlah_stok = validasi_input("Masukkan jumlah stok barang: ", int)

        kode_barang = generate_kode_barang()  # Menghasilkan kode barang baru
        barang = Barang(kode_barang, nama_barang, harga, jumlah_stok)  # Membuat objek Barang
        barang_dict[kode_barang] = barang  # Menyimpan objek Barang ke dalam dictionary
        print(Style.BRIGHT +
              Fore.GREEN +
              f"Barang {nama_barang} dengan kode {kode_barang} berhasil ditambahkan." +
              Style.RESET_ALL)
        break

# Fungsi untuk mencari barang berdasarkan kode atau nama
def cari_barang():
    while True:
        print(Style.BRIGHT +
              Fore.CYAN +
              "\nPilih cara pencarian:" +
              Style.RESET_ALL)
        print("1. Cari berdasarkan Kode Barang")
        print("2. Cari berdasarkan Nama Barang")
        print("3. Kembali ke Menu Utama" +
              Style.RESET_ALL)
        pilihan = input("Masukkan pilihan (1/2/3): ").strip()

        if pilihan == "1":
            kode = input("Masukkan kode barang yang ingin dicari: ").strip()
            barang = barang_dict.get(kode)  # Mencari barang berdasarkan kode
            if barang:
                print(Style.BRIGHT +
                      Fore.YELLOW +
                      barang.info_barang() +
                      Style.RESET_ALL)
            else:
                print(Style.BRIGHT +
                      Fore.RED +
                      "Barang dengan kode tersebut tidak ditemukan." +
                      Style.RESET_ALL)
        elif pilihan == "2":
            nama = input("Masukkan nama barang yang ingin dicari: ").strip().lower()
            # Mencari barang berdasarkan nama
            found = [barang.info_barang() for barang in barang_dict.values() if nama in barang.nama_barang.lower()]
            if found:
                for info in found:
                    print(Style.BRIGHT +
                          Fore.YELLOW +
                          info +
                          Style.RESET_ALL)
            else:
                print(Style.BRIGHT +
                      Fore.RED +
                      "Barang dengan nama tersebut tidak ditemukan." +
                      Style.RESET_ALL)
        elif pilihan == "3":
            break
        else:
            print(Style.BRIGHT +
                  Fore.RED +
                  "Pilihan tidak valid. Silakan coba lagi." +
                  Style.RESET_ALL)

# Fungsi untuk menampilkan semua barang yang ada dalam inventaris
def tampilkan_semua_barang():
    if not barang_dict:
        print(Style.BRIGHT +
              Fore.RED +
              "Inventaris kosong. Tidak ada barang yang ditampilkan." +
              Style.RESET_ALL)
    else:
        print(Style.BRIGHT +
              Fore.CYAN +
              "\nDaftar Semua Barang:" +
              Style.RESET_ALL)
        for barang in barang_dict.values():
            print(Style.BRIGHT +
                  Fore.YELLOW +
                  barang.info_barang() +
                  Style.RESET_ALL)

# Fungsi untuk memperbarui stok barang berdasarkan kode
def update_stok_barang():
    while True:
        kode_barang = input("Masukkan kode barang yang ingin diperbarui stoknya (ketik 'kembali' untuk kembali ke menu): ").strip()
        if kode_barang.lower() == 'kembali':
            break
        barang = barang_dict.get(kode_barang)  # Mencari barang berdasarkan kode
        if barang:
            jumlah = validasi_input("Masukkan jumlah stok yang akan ditambahkan atau dikurangi, (-) untuk mengurangi: ", int)
            barang.update_stok(jumlah)  # Memperbarui stok barang
        else:
            print(Style.BRIGHT +
                  Fore.RED +
                  "Barang dengan kode tersebut tidak ditemukan." +
                  Style.RESET_ALL)

# Fungsi untuk menampilkan menu utama dan menangani pilihan pengguna
def menu():
    while True:
        print(Style.BRIGHT +
              Fore.CYAN +
              "\nMenu:" +
              Style.RESET_ALL)
        print("1. Tampilkan Semua Barang")
        print("2. Tambah Barang")
        print("3. Cari Barang")
        print("4. Update Stok Barang")
        print("5. Keluar")

        pilihan = input("Pilih menu (1/2/3/4/5): ").strip()

        if pilihan == "1":
            tampilkan_semua_barang()  # Menampilkan semua barang
        elif pilihan == "2":
            tambah_barang()  # Menambah barang baru
        elif pilihan == "3":
            cari_barang()  # Mencari barang
        elif pilihan == "4":
            update_stok_barang()  # Memperbarui stok barang
        elif pilihan == "5":
            print(Style.BRIGHT +
                  Fore.GREEN +
                  "Terima kasih telah menggunakan program ini." +
                  Style.RESET_ALL)
            break
        else:
            print(Style.BRIGHT +
                  Fore.RED +
                  "Pilihan tidak valid. Silakan coba lagi." +
                  Style.RESET_ALL)

# Memanggil fungsi menu untuk memulai program
menu()
```
